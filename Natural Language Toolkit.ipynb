```python
import nltk
# nltk.download()
%matplotlib inline
```

```python
sentence = "At eight o'clock on Thursday morning. Arthur didn't feel very good."
```

```python
tokens = nltk.word_tokenize(sentence)
print(tokens)
print(len(tokens))
```

```python
sent_tokens = nltk.sent_tokenize(sentence)
print(sent_tokens)
print(len(sent_tokens))
```

```python
e_words = ["wait", "waiting", "waited", "waits"]
ps = nltk.PorterStemmer()
for w in e_words:
    rootWord=ps.stem(w)
    print(rootWord)
```

```python
sentence="Hello Guru99, You have to build a very good site and I love visiting your site."
words = nltk.word_tokenize(sentence)
ps = nltk.PorterStemmer()
for w in words:
    rootWord=ps.stem(w)
    print(rootWord)
```

```python
import re
import pandas as pd
text = "Guru99 is one of the best sites to learn WEB, SAP, Ethical Hacking and much more online."
lower_case = text.lower()
words = nltk.word_tokenize(lower_case)
nonPunct = re.compile('.*[A-Za-z0-9].*')  # must contain a letter or digit
filtered = [w for w in words if nonPunct.match(w)]
counts = nltk.Counter(filtered)
print(counts)
```

```python
a = "Guru99 is the site where you can find the best tutorials for Software Testing     Tutorial, SAP Course for Beginners. Java Tutorial for Beginners and much more. Please     visit the site guru99.com and much more."
words = nltk.tokenize.word_tokenize(a)
fd = nltk.FreqDist(words)
fd.plot()
```

```python
stop_words=set(nltk.corpus.stopwords.words("english"))
print(stop_words)
```

```python
text = """Hello Mr. Smith, how are you doing today? The weather is great, and city is awesome.
The sky is pinkish-blue. You shouldn't eat cardboard"""
sentence = text.lower()
tokenized_sent = nltk.word_tokenize(sentence)
filtered_sent=[]
for w in tokenized_sent:
    if w not in stop_words:
        filtered_sent.append(w)
print("Tokenized Sentence:",tokenized_sent)
print("Filterd Sentence:",filtered_sent)
```

```python
import pandas as pd 
import numpy as np 
```

```python
df = pd.DataFrame({
    "text": [['ill', 'be', 'the'], ['wow', 'so'], ['their'],['ive', 'never'],['easily'],['employees', 'were'], ['great', 'tanning']],
    "stars": [4.0, 4.0, 5.0, 1.0, 5.0, 2.0, 5.0]
})
df.head(n=7)
```

```python
df["art_id"] = df.index.values + 1
df.head(n=7)
```

```python
df_explode = df.explode('text')
df_explode.head(n=13)
```

```python
df_dummies = pd.get_dummies(df_explode, prefix = 'w', columns=['text'])
df_dummies.head(n=13)
```

```python
df_groupby = df_dummies.groupby(["stars", "art_id"]).sum().reset_index()
df_words = df_groupby.drop(["art_id"], axis=1)
df_words.head(n=7)
```
