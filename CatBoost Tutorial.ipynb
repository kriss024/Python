```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from catboost import CatBoostClassifier, Pool
from catboost.datasets import adult
from IPython.display import display
```

```python
adult_train, adult_test = adult()
```

```python
display(adult_train)
```

```python
display(adult_test)
```

```python
adult_train['income-over-50K'] = adult_train['income'].apply(lambda x: 1 if x == '>50K' else 0)
adult_train.drop(['income'], axis=1, inplace=True)
```

```python
display(adult_train)
```

```python
adult_test['income-over-50K'] = adult_test['income'].apply(lambda x: 1 if x == '>50K' else 0)
adult_test.drop(['income'], axis=1, inplace=True)
```

```python
display(adult_test)
```

```python
train_data = adult_train[['age', 'fnlwgt', 'education-num', 'capital-gain', 'capital-loss', 'hours-per-week']]
train_labels = adult_train[['income-over-50K']]

# In the first step we will split the data in training and remaining dataset
X_train, X_rem, y_train, y_rem = train_test_split(train_data, train_labels, 
                                                    stratify=train_labels, test_size=0.4, random_state=42)

# Now since we want the valid and test size to be equal (10% each of overall data). 
# we have to define valid_size=0.5 (that is 50% of remaining data)
test_size = 0.5
X_valid, X_test, y_valid, y_test = train_test_split(X_rem,y_rem, test_size=test_size)

X_test = adult_test[['age', 'fnlwgt', 'education-num', 'capital-gain', 'capital-loss', 'hours-per-week']]
y_test = adult_test[['income-over-50K']]
```

```python
display(X_train)
```

```python
display(y_test)
```

```python
train_pool = Pool(X_train, y_train)
valid_pool = Pool(X_valid, y_valid)
test_pool = Pool(X_test, y_test)
```

```python
model = CatBoostClassifier(learning_rate=0.03, eval_metric='AUC')
```

```python
# train the model
model.fit(train_pool, eval_set=valid_pool, verbose=False)
```

```python
print(model.get_best_iteration())
print(model.get_best_score())
print(model.get_feature_importance(train_pool))
```

```python
model.save_model("model.cbm")
del model

model = CatBoostClassifier()
model.load_model("model.cbm")
```

```python
# make the prediction using the resulting model
preds_class = model.predict(test_pool)
preds_proba = model.predict_proba(test_pool)
print("class = ", preds_class)
# print("proba = ", preds_proba)
```

```python
pd.crosstab(test_pool.get_label(), preds_class, 
            rownames=['Original class'], colnames=['Predicted class'], dropna=False)
```
